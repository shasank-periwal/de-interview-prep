```python
bool isKthBitSet(int n, int k) {
    return (n & (1 << (k - 1))) != 0;
}
```

```python
int turnOffKthBit(int n, int k) {
    return n & ~(1 << (k - 1));
}
```

```python
def inbinary(num):
    num_base3 = ""
    while(num):
        num_base3 = str(num%2) + num_base3
        num = num // 2
    # print(num_base3)
    return num_base3

inbinary(4)
inbinary(8)
inbinary(27)
inbinary(64)
inbinary(81)
```

```python
from collections import defaultdict
arr = [20, 2, 7, 11, 16, 9, 4 , 5]
target = 9

res = []
mp = defaultdict(int) 
mp[0] = len(arr)

for idx, num in enumerate(arr): 
    if mp[num] != 0:
        res.append([mp[num], idx])
    elif num - target == 0:
        res.append([idx])
    mp[target - num] = idx

print(res)

# 20 
# temp = -11, 
```

```python
arr = [20, 2, 7, 11, 16, 9, 4 , 5]

def insert(arr, ele):
    if len(arr) == 0 or arr[len(arr)-1]<=ele:
        arr.append(ele)
        return

    tmp = arr.pop()
    insert(arr, ele)
    arr.append(tmp)


def sort(arr):
    if len(arr)<=1:
        return 
    
    ele = arr.pop()

    sort(arr)

    insert(arr, ele)

print(arr)
sort(arr)
print(arr)


```

```python
import json

datas = {
    "/web/leetcode/contest": 435,
    "/web/leetcode/discuss": "Zomato",
    "/web/leetcode/premium/amount": 15000,
    "/leetcode/explore": "interviewExp",
    "/leetcode/store": "premium"
}

nested_dict = {}

for key, value in datas.items():
    splitted = [part for part in key.split("/") if part]
    current_lvl = nested_dict
    
    for val in splitted[:-1]:
        if val not in current_lvl.keys():
            current_lvl[val] = {}
        current_lvl = current_lvl[val]
    
    current_lvl[splitted[-1]] = value


print(json.dumps(nested_dict))
```

```python
def findGCD(a, b):
    if a == 0:
        return b
    return findGCD(b % a, a)

def caller(a,b):
    gcd = findGCD(a,b)
    print("GCD: ", gcd)
    print("LCM: ", int((a*b)/gcd))

caller(23,20677)
# [2009,899,20677,825]
```

```python
# Write a program to find all possible combinations of a string in Python.

def allComb(st, idx: int = 0, ln: int = 0):
    if idx >= ln:
        print("".join(st))
        return

    for i in range(idx, ln):
        st[idx], st[i] = st[i], st[idx]
        allComb(st, idx+1, ln)
        st[idx], st[i] = st[i], st[idx]
        
    return

allComb(st=list('hank'), idx=0, ln=len('hank'))
```
