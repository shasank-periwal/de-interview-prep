```python
ans = ""
res = ""
prev = None
count = 1
for i in range(len(ans)):
    print(i, res, prev, ans[i])
    if prev and prev == ans[i]:
        count += 1
    elif (prev and prev != ans[i]):
        res += str(count) + prev
        count = 1
        prev = ans[i]
    else:
        prev = ans[i]
res += str(count) + prev
print(res)
```
