```python
import pyspark
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
```

```python
conf = pyspark.SparkConf().setAppName("wordcount").setMaster("local")
sc = SparkSession.builder.config(conf = conf).getOrCreate()
```

```python
# df = sc.read.csv("./assets/currency.csv", header=True, inferSchema=True)
text_file = sc.read.text("./assets/words.txt").rdd.map(lambda row:row[0])
print(text_file.collect())
```

```python
# df.show()
# df.withColumn('word', explode(split(col('value'), ' ')))
#     .groupBy('word')
#     .count()
    # .sort('count', ascending=False)
cnt = text_file.flatMap(lambda line: line.split(" ")).map(lambda word: (word, 1)).reduceByKey(lambda x, y: x+y)
print(cnt.collect())
```
